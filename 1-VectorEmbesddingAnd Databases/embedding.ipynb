### Creating Your First Embeddings

```python
### Huggingface and OenAI Models
# import os
# os.environ["HF_HUB_DISABLE_SSL_VERIFICATION"] = "1"

from langchain_huggingface import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(
    model_name = "sentence-transformers/all-MiniLM-L6-v2"
)
embeddings
```

```python
# create your first embedding
text = "Hello, I am leaning about embeddings!"

embedding = embeddings.embed_query(text)
print(f"Text: {text}")
print(f"Embedding lenght: {len(embedding)}")
print(embedding)
```

```python
sentences = [
    "The cat sat on the mat",
    "A feline rested on the rug",
    "The dog played in the yard",
    "I love programming in python",
    "Python is my favorite programming language"
]

embedding_sentence = embeddings.embed_documents(sentences)

print(embedding_sentence)
```
